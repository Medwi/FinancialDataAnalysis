## Determining if distributions are different (T-Test and Mann-Whitney U Test)
The averages vary slightly, lets run a T-Test to see if this is by chance or significant. Comparing the stocks that meet our EPS criteria with the rest of our sample. For this test I will use T-test and the Mann Whiteney U. I will use an alpha of .05 for these test. 

**T-Test** : Determines if the means of two groups are equal to each other. 

**Mann Whitney U** : The Mann-Whitney U test is a nonparametric statistical significance test for determining whether two independent samples were drawn from a population with the same distribution. If this test concludes that the stocks with the EPS filter came from the same population as the non-EPS stocks than this is strong evidence to suggest that the EPS criteria does not make a difference.

***Null Hypothesis*** - Stocks Price Percentage change from stocks which meet the EPS Criteria are the same as stocks which don't

***Alternative Hypothesis*** - Stocks Price Percentage change from stocks which meet the EPS Criteria is different than stocks which don't

In both cases if the p value < alpha we reject the Null Hypothesis

```python
print(stats.ttest_ind(stocks[EPS_filter == False]['Pricep target'].dropna(),stocks[EPS_filter]['Pricep target'].dropna(), equal_var=False))
print(stats.mannwhitneyu(stocks[EPS_filter == False]['Pricep target'].dropna(),stocks[EPS_filter]['Pricep target'].dropna()))
print(stats.f_oneway(stocks[EPS_filter == False]['Pricep target'].dropna(),stocks[EPS_filter]['Pricep target'].dropna()))
```

Our P-Values are .02 for the T-Test and .006 for the Mann Whitney U which are both less than our alpha of .05. This tells us we can reject the NULL hypothesis and conclude that the means between the stocks that meet the EPS criteria and those which don't are different.

## Testing samples
Our sample sizes are vastly different between the distributions we are comparing, i.e 57,000 for the non-EPS and 384 for the EPS. So to check to see if we get drastically different values comparing the distributions if we compared same size samples I run the t-test again 1000 times using 1000 different samples. 

```python
p_val = []
for r in range(1000):
    t, p = stats.ttest_ind(stocks[EPS_filter == False]['Pricep target'].dropna().sample(384,random_state=r),stocks[EPS_filter]['Pricep target'].dropna(), equal_var=False)
    p_val.append(p)
fig = plt.figure(figsize=(10,5))
plt.hist(p_val,bins=20)
plt.xticks(np.arange(0,1,.05))
plt.title('P Values for 1000 T-Test(Sample size 384)')
plt.show()
```

Although under .05 p-value was the most common value which would sugguest to reject the NULL hypothesis a lot of variability  existed depending on our sample. The Whitney Mann U and T-Test used above are non-dependent on sample size, but we can see here it does make a difference if we took a equal sample approach. This is likely due to how much variability of stock price percentage change exists among our data.

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard Deviation:", scores.std())
```

## Running a Quick Random Forest Regressor

```python
f_reg = RandomForestRegressor()
f_reg.fit(X_train, Y_train)
scores = cross_val_score(f_reg, X_train, Y_train, scoring="neg_mean_squared_error", cv=10)
f_reg_scores = np.sqrt(-scores)
display_scores(f_reg_scores)
```

The quick Random Forest Regressor model had a mean error of 14.3% with a std of 2% . This is a pretty high error considering most of our price data falls between -4% and 4%. I was not expecting a very well performing model considering how noisy our data is. 

## Randomized Search to find a better model

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf_random = RandomizedSearchCV(estimator = f_reg, param_distributions = random_grid, n_iter = 100, cv = 5, verbose=2, random_state=42, n_jobs = -1)
rf_random.fit(X_train, Y_train)
```

```python
rf_best = rf_random.best_params_
rf_best_features = rf_random.best_estimator_
```

```python
scores = cross_val_score(rf_best_features, X_train, Y_train, scoring="neg_mean_squared_error", cv=10)
rf_best_features_scores = np.sqrt(-scores)
display_scores(rf_best_features_scores)
```

```python
final_predictions = rf_best_features.predict(X_test)
final_mse = mean_squared_error(Y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
final_rmse
```

Our model returned by randomized search had the following accuracy:

- Training Set : 13.4%
- Test Set : 13.8%

This is a minor improvement on our previous model and doesn't appear to be overfitting or underfitting our data terribly considering how close our training and test set accuracy is. I will use this model to see how the feature importance compares to our correlation above.

```python
feature_rank = sorted(zip(rf_best_features.feature_importances_, X_train.columns), reverse=False)
fig=plt.figure(figsize=(10,15))
plt.barh(range(len(feature_rank)), [i[0] for i in feature_rank],align='center')
plt.yticks(range(len(feature_rank)), [i[1] for i in feature_rank],fontsize=10)
plt.title('Random Forest Feature Importance')
plt.show()
```

Like the correlation I ran above, P/B Ratio change by quarter is the overwhelming the best predictor of stock price change.
